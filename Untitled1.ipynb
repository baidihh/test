```python
from pyecharts import Line
import pandas as pd
data = pd.read_csv('./data/data.csv')
data
datax = ['星期一','星期二','星期三','星期四','星期五','星期六','星期日']
```

```python
from pyecharts import Line
line = Line("死亡与治愈人数变化图",title_pos = 'left')
line.add('治愈人数',datax,data['治愈'],
         is_step=False,
         xaxis_name ='星期几',
         yaxis_name ='人数',
         xaxis_name_pos='end' ,
         yaxis_name_pos='end' ,
         xaxis_interval  = 0,
         yaxis_formatter = '/人',
#          yaxis_range=(data['治愈'].min,data['治愈'].max),
        yaxis_range=[20,40],
         line_type = 'solid',
         line_color='blue',
         line_width =3,
         is_label_show=True,
#          is_splitline_show =False,
         mark_point =['min','max']
        )
line.add('死亡人数',datax,data['死亡'],
         is_step=False,
         xaxis_name ='星期几',
         yaxis_name ='人数',
         xaxis_name_pos='end' ,
         yaxis_name_pos='end' ,
         xaxis_interval  = 0,
         yaxis_formatter = '/人',
         line_type = '++',
         line_color='red',
         line_width =3,
         is_label_show=True,
         mark_point =['min','max']
        )
line
```
